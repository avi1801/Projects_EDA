### The govt should implement the Lockdown in some states as per the current suitation the respective states, we have define Lockdown into 3 types:-

### Lockdown 1: Complete Lockdown
* Maharashtra, Kerala, Karnataka, Tamil Nadu, Andhra Pradesh, Uttar Pradesh, West Bengal, Delhi
### Lockdown 2: Wear mask and only essential services are allowed to open
* Chhattisgarh, Rajasthan, Odisha, Gujarat, Madhya Pradesh, Haryana, Bihar, Telangana,Punjab,Assam
### Lockdown 3: Should wear the mask
* Jharkhand, Uttrakhand, Jammu and Kashmir, Himachal Pradesh, Goa, Pondicherry, Manipur, Tripura, Chhattisgarh, Meghalaya, Arunachal Pradesh, Nagaland, Mizoram, Sikkim, Ladakh, Dadar Nagar Haveli Daman Diu, Lakshadweep, Andaman Nicobar

```python
df
```

```python
df["Lockdown_Status"]=df["Total Cases"].apply(lambda x: "Lockdown 1" if ((x>=1434780) & (x<=6122893)) else ("Lockdown 2" if ((x<1434780) & (x>=524556)) else "Lockdown 3"))
```

```python
df
```

### We have created one column Name Lockdown_Status, just to help the govt to easily identity the Lockdown Status of all the state  

```python
df.head()
```

```python
df["Recovered Cases"]=df["Total Cases"]-(df["Active"]+df["Deaths"])
df
```

### We have created one column name Recovered Cases, just to identity how many cases got recovered 
* Recovered Cases=(Active + Death) - Total Cases

```python
df["Recovered Cases"].sum()
```

### Observation:- Total sum od Recovered cases in all States/UTs is 29812873.
### It seems that many of the people recovered from Covid-19


```python
df.columns
```

```python
df1=df[["State/UTs","Total Cases",'Active', 'Deaths',"Recovered Cases","Active Ratio (%)",'Death Ratio (%)','Lockdown_Status']]
```

```python
df1
```

### We have Reorder the column Recovered Cases

```python
df1["Recovered Ratio(%)"]=round(df1["Recovered Cases"]*100/df1["Total Cases"],2)
```

```python
df1
```

### We have created one more column name Recovered Ratio(%) and also Reordered that column adjacent to Recovered Status

```python
df1=df1[["State/UTs","Total Cases",'Active', 'Deaths',"Recovered Cases","Active Ratio (%)",'Death Ratio (%)',"Recovered Ratio(%)",'Lockdown_Status']]
```

```python
df1
```

```python
d=df1.sort_values(by="Deaths",ascending=False)
plt.figure(figsize=(12,10))
sns.barplot(x="State/UTs",y="Deaths",data=d)
plt.xticks(rotation=90,fontweight="bold",fontsize=10);
```

```python
df1["Death Ratio (%)"].agg(["max","min"])
```

```python
df1["Fund_Allocation_Status"]=df1["Death Ratio (%)"].apply(lambda x:"Fund Allocated" if x>1.5 else "No")
```

### In above we created one column name Fund_Allocation_Status, to help the govt to identify the states who need some sort Funds 
### So have give the Funds by calculating the Death Ratio of the states where the ratio is less than 1.5
